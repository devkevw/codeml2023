```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier #decision trees
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
# step 1: load data
enviro_data = pd.read_csv('data_participant_ECCC.csv')
enviro_data 
```

```python
# step 2: clean data
df_thun = enviro_data.drop(columns = ['year', 'month', 'day', 'hour', 'start_time', 'latitude', 'longitude', 'event', 'y_hail', 'y_severe'])
df_hail = enviro_data.drop(columns = ['year', 'month', 'day', 'hour', 'start_time', 'latitude', 'longitude', 'event', 'y_thunderstorm', 'y_severe'])
df_sev = enviro_data.drop(columns = ['year', 'month', 'day', 'hour', 'start_time', 'latitude', 'longitude', 'event', 'y_thunderstorm', 'y_hail'])
```

```python
# step 3: split the data into training/test sets

# find prediction data
prediction_df_thun = df_thun[df_thun['y_thunderstorm'].isna()]
prediction_df_hail = df_hail[df_hail['y_hail'].isna()]
prediction_df_sev = df_sev[df_sev['y_severe'].isna()]

# find remaining dataset
dataset_thun = df_thun.dropna(how='any')
dataset_hail = df_hail.dropna(how='any')
dataset_sev = df_sev.dropna(how='any')

# split into training and testing
X_thun = dataset_thun.drop(columns = ['y_thunderstorm'])
y_thun = dataset_thun['y_thunderstorm']

X_hail = dataset_hail.drop(columns = ['y_hail'])
y_hail = dataset_hail['y_hail']

X_sev = dataset_sev.drop(columns = ['y_severe'])
y_sev = dataset_sev['y_severe']


# convert true false to 0, 1
label_encoder = LabelEncoder()
y_thun = label_encoder.fit_transform(y_thun)
y_hail = label_encoder.fit_transform(y_hail)
y_sev = label_encoder.fit_transform(y_sev)

# split into 80% traiing, 20% testing
X_train_thun, X_test_thun, y_train_thun, y_test_thun = train_test_split(X_thun, y_thun, test_size = 0.2)
X_train_hail, X_test_hail, y_train_hail, y_test_hail = train_test_split(X_hail, y_hail, test_size = 0.2)
X_train_sev, X_test_sev, y_train_sev, y_test_sev = train_test_split(X_sev, y_sev, test_size = 0.2)
```

```python
# step 4 & 5: create model (decision tree) & train model
model_thun = DecisionTreeClassifier()
model_thun.fit(X_train_thun.drop(columns=['ID']), y_train_thun)

model_hail = DecisionTreeClassifier()
model_hail.fit(X_train_hail.drop(columns=['ID']), y_train_hail)

model_sev = DecisionTreeClassifier()
model_sev.fit(X_train_sev.drop(columns=['ID']), y_train_sev)
```

```python
# step 6: make predictions and check accuracy to tune model
predictions_thun = model_thun.predict(X_test_thun.drop(columns=['ID']))
accuracy_score(y_test_thun, predictions_thun)

predictions_hail = model_hail.predict(X_test_hail.drop(columns=['ID']))
accuracy_score(y_test_hail, predictions_hail)

predictions_sev = model_sev.predict(X_test_sev.drop(columns=['ID']))
accuracy_score(y_test_sev, predictions_sev)
```

```python
# step 7: use the model
act_predictions_thun = model_thun.predict(prediction_df_thun.drop(columns = ['y_thunderstorm', 'ID']))
act_predictions_hail = model_hail.predict(prediction_df_hail.drop(columns = ['y_hail', 'ID']))
act_predictions_sev = model_sev.predict(prediction_df_sev.drop(columns = ['y_severe', 'ID']))

# convert actual predictions back to True, False
pred_tf_thun = act_predictions_thun.astype(bool)
pred_tf_hail = act_predictions_hail.astype(bool)
pred_tf_sev = act_predictions_sev.astype(bool)
```

```python
# filter for ID, predictions
prediction_df_thun.loc[:,'y_thunderstorm'] = pred_tf_thun
to_add_thun = prediction_df_thun[['ID', 'y_thunderstorm']]

prediction_df_hail.loc[:,'y_hail'] = pred_tf_hail
to_add_hail= prediction_df_hail[['ID', 'y_hail']]

prediction_df_sev.loc[:,'y_severe'] = pred_tf_sev
to_add_sev= prediction_df_sev[['ID', 'y_severe']]
```

```python
# replace original data with predicition, matching using ID
enviro_data.set_index('ID', inplace=True)
to_add_thun.set_index('ID', inplace=True)
to_add_hail.set_index('ID', inplace=True)
to_add_sev.set_index('ID', inplace=True)

enviro_data.update(to_add_thun)
enviro_data.update(to_add_hail)
enviro_data.update(to_add_sev)

enviro_data.reset_index(inplace=True)
```

```python
# save to output csv file
enviro_data.to_csv('output_environCan', index=False)
```
